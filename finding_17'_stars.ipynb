## This notebook is to find stars under 9' from 0855 -- our promising candidate

```python
import pandas as pd
import numpy as np

from PyAstronomy import pyasl
```

```python
df = pd.read_csv('irsa_catalog_search_results_tbl.csv')
```

```python
#filter all stars dimmer than j-band mag of 9
under_09_mag = df['j_m'] < 9
df = df[under_09_mag]
```

```python
#check distances for each star

#coords used for 2MASS
ra = 133.770952
dec = -7.243109

distance_array = list()
for i in range(len(df)):
    theta = pyasl.getAngDist(ra, dec, list(df['ra'])[i], list(df['dec'])[i])
    
    #convert from radians to arcminutes
    theta *= (2 * np.pi * (1 / 360)) * 206264 * (1 / 60)
    
    distance_array.append(theta)
```

```python
#now make a df including the ra, dec, mag of background star
df_bright = pd.DataFrame(_ , columns=['ra', 'dec', 'sep', 'mag'])

df_bright['ra'] = df['ra']
df_bright['dec'] = df['dec']
df_bright['sep'] = distance_array
df_bright['mag'] = df['j_m']
```
